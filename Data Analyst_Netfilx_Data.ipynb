```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
film = pd.read_csv('netflix_data.csv')
film.head()
```

```python
film.shape
```

```python
film.info()
```

```python
film.describe(include='all')
```

```python
film.duplicated().value_counts()
```

```python
film.isna().sum().sort_values(ascending = False)
```

```python
film['director'].fillna('Others',inplace = True)
film['country'].fillna('Not Avaliable',inplace = True)
film['cast'].fillna('Others',inplace = True)
film['date_added'].fillna('Not Available',inplace = True)
film['rating'].fillna('Others',inplace = True)
film['duration'].fillna('Not Available',inplace = True)
```

```python
film.info()
```

```python
top_release_years = (
    film['release_year']
    .value_counts()
    .head(15)
)
top_release_years
```

```python
plt.figure(figsize=(20, 5))
sns.barplot(x=top_release_years.index, y=top_release_years.values, color='yellow')
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Proportion of Production per Year')
plt.show()
```

```python
show = film["director"].value_counts().reset_index()[1:11]
show
```

```python
plt.figure(figsize=(20, 5))
sns.barplot(x = 'director',y = 'index',color = 'purple',data = show)
plt.title('Movie Director and Total Films',fontsize=15)
plt.show()
```

```python
show = film["cast"].value_counts().reset_index()[1:11]
show
```

```python
sns.barplot(x = 'cast',y = 'index',color = 'purple',data = show)
plt.title('Cast and Total Films',fontsize=15)
plt.xticks(rotation = 90)
plt.show()
```

```python
labels = ['Movie', 'TV show']
size = film['type'].value_counts()
colors = ['red','green']
explode = [0, 0.1]
plt.rcParams['figure.figsize'] = (5, 5)
ax = plt.pie(size,labels=labels,colors=colors, explode=explode, shadow=True, startangle=120, autopct='%1.1f%%')
plt.title('Comparison Type', fontsize=10)
plt.legend()
plt.show()
```

```python
show1 = film[film['type'] == "Movie"]
show2 = film[film['type'] == "TV Show"]
```

```python
plt.figure(figsize=(20, 5))
movie = show1.groupby('country')['show_id'].count().sort_values(ascending = False).head(10)
movie.plot(kind = 'barh', title = 'Highest Number of Movies Viewers by Country', color = 'red')
plt.show()
```

```python
plt.figure(figsize=(20, 5))
Tv_Show = show2.groupby('country')['show_id'].count().sort_values(ascending = False).head(10)
Tv_Show.plot(kind = 'barh', title = 'Highest Number of TV Show Viewers by Country', color = 'green')
plt.show()
```

```python

rating_count = film['rating'].value_counts()

x = rating_count.index
y = rating_count

plt.figure(figsize=(20,5))
plt.bar(x,y,color='orange')
plt.ylabel('Amount of contents')
plt.title('Number of Content Based on Rating')
plt.show()
```
